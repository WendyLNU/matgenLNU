MP网站的工作流的主要功能：
1、灵活的Python API ，强大的命令行界面和可以监控工作流状态的内置WEB服务器。
2、后端数据库（MongDB）允许我们添加，删除和搜素工作流的状态。
3、检测失败的作业（软故障和硬故障），并且根据需求重新运行它们。
4、多种执行模式：（1）直接在多核机器上运行    （2）  通过队列，在一台机器或者多台机器上运行。分配优先级和作业运行的位置。
5、支持动态工作流程：根据执行期间发生的情况自行修改或者创建新的工作流程。
6、子工作流级别的自动重复处理：当两个工作流同时运行唯一的部分时，跳过两个工作流之间的重复部分。
7、内置任务，用于创建模板化输入、运行脚本以及将文件复制到远程计算机。
8、在执行期间远程跟踪输出文件的状态。
9、将许多小型作业打包成一个大型作业（例如：在100个内核上并行运行100个串行工作流程）。
10、支持多种排队系统，如PBS/Torque, Sun Grid Engine, SLURM, and IBM LoadLeveler。
